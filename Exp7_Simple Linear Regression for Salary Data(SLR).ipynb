# Independent and Dependent Variables

```python
x=df.drop('Salary',axis=1)
```

```python
x.head()
```

```python
y=df.Salary
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train, y_test = train_test_split(x, y, random_state=0,test_size=0.30)
```

```python
print(x_train.shape)
```

```python
x_test.shape
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(df['Salary'], df['YearsExperience'], marker='o')
plt.xlabel("Salary")
plt.ylabel("YearsExperience")
plt.show()
```

# Model Fitting

```python
from sklearn.linear_model import LinearRegression
LR=LinearRegression()
LR.fit(x_train,y_train)
```

```python
#Assigning Coefficient (Slope) to m
m=LR.coef_
```

```python
print("Coefficient  :" , m)
```

```python
#Assigning Y-intercept to a
c=LR.intercept_
```

```python
print("Intercept : ", c)
```

# Evaluation Metrics

```python
from sklearn import metrics
```

```python
Accuracy = LR.score(x_test, y_test)
Accuracy
```

# Conclusion :
               The simple linear regression model shows a clear linear relationship between the variables, allowing us to predict outcomes based on this trend with reasonable accuracy.
